```python
# test['probability'].plot(kind='hist')
probable_probs = test.groupby('category_level_1_1')[['all_probability', 'cat1_probability_1', 'cat1_probability_2']].mean()
```

```python
test['fixed_cat1_probability_1'] = test['cat1_probability_1']
test['fixed_cat1_probability_2'] = test['cat1_probability_2']
```

```python
for model in ['cat1_probability_1', 'cat1_probability_2']:
    for cat in probable_probs.index:
        # target_prob = 
        p1_before = probable_probs.loc[cat]['all_probability']
        p1_after = probable_probs.loc[cat][model]
        p0_before = 1 - p1_before
        p0_after = 1 - p1_after
        p1_actual = test[test['category_level_1_1'] == cat][model]
        p1_adj = (
            p1_actual * p1_before * p0_after) / (
            p1_actual * p1_before * p0_after + (1 - p1_actual) * p0_before * p1_after
        )
        test.loc[test['category_level_1_1'] == cat, 'fixed_' + model] = p1_adj
        # print(p1_actual, p1_adj)
        # print(p1_actual.mean(), p1_adj.mean())
        # p1_actual.plot(kind='hist')
        # p1_adj.plot(kind='hist')
        # display(probable_probs)
        
        # print(target_prob)
        # display(test[test['category_level_1_1'] == cat])
```
